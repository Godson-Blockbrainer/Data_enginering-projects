```python
# change the data type of the churn column from boolean to int64 and check the dataframe again

df['Churn']= df.Churn.astype('int64')
df

```

```python
# get the distribution of the churn by counting how many churned and how many did not
churn_distribution = df.Churn.value_counts()
print(churn_distribution)
```

```python
# plot the count from the result above
churn_distribution = df['Churn'].value_counts()

churn_distribution.plot(kind='bar', color=['skyblue','navy'])
plt.title('Churn Distribution')
plt.xlabel('Churned')
plt.ylabel('Count')

```

```python
# what is the proportion of churned users in the dataframe
churned_users =df.Churn.sum()
total_users =len(df)

churned_proportion= churned_users/total_users
print("we have",churned_proportion*100 ,"of churned customers in the dataset")
```

```python
# How much time (on average) do churned users spend on the phone during daytime?
Average_daytime_by_churned_users =df[df['Churn']==1]['Total day minutes'].mean()
print(Average_daytime_by_churned_users)
```

```python
# What is the maximum length of international calls among loyal users (Churn == 0) who do not have an international plan?
maximum_international_call_length = df[(df['Churn'] == 0) & (df['International plan'] == 'No')]['Total intl minutes'].max()
print(maximum_international_call_length)

# so we filtered first condition which was to set our churn to egte a more filtered column
#of out loyal users , and also we need to filter the international call column and choose only those 
# assigned to =='No' afterwards the last filters for the new table generated where this two new clumans are true
```

```python
# give the values of the first five rows in the first three columns
first_five_rows = df.iloc[:5,:3]
first_five_rows
```

```python
# select all the states that starts with W

states_with_W = df[df['State'].str.startswith('W')]
states_with_W.head(5)
```

```python
# using the map function, replace Yes and No in the International plan column to True and False
replace_dict = {'Yes': True, 'No': False}
df['International plan'] = df['International plan'].map(replace_dict)
df
```

```python
# we want to understand the total calls that have been made by customers. Total_calls is calculated as
# the sum of Total dat calls, Total eve calls, Total night calls, Total intl calls.

df['Totalcalls'] = df['Total day calls'] + df['Total eve calls'] + df['Total night calls'] + df['Total intl calls']
df.head(5)
```
